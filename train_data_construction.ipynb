```python
import pandas as pd
import json
```

```python
with open("/content/唐诗三百首.json") as file:
  data = json.load(file)
  print(data[0])
```

```python
normal_paragraphs = []
for each in data:
  normal_paragraphs.append(each["paragraphs"])
print(normal_paragraphs[0:20])
```

```python
final_paragraphs = []
for each in normal_paragraphs:
  final = ""
  for every in each:
    final = final+every
  final_paragraphs.append(final)
print(final_paragraphs[0:20])
```

```python
title = []
for each in data:
  title.append(each['title'])
print(title[0:20])
```

```python
from nltk.probability import FreqDist
token_numbers = []
for each in final_paragraphs:
  fdist = FreqDist(each)
  all = fdist.N()
  token_numbers.append(all)
print(token_numbers[0:20])
```
